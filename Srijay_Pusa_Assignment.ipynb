```
import pandas as pd

url = "https://drive.google.com/file/d/1Mlu9r3afSGz5_02Y3wDJoQLpXszQXaIU/view"
file_id = url.split('/')[-2]
download_url = f"https://drive.google.com/uc?export=download&id={file_id}"

df = pd.read_csv(download_url, index_col='order_id')
print("First 5 rows:\n", df.head())
print("\nLast 5 rows:\n", df.tail())
```

```
print("\nDataset shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum())
print("\nData types:\n", df.dtypes)
df['order_date'] = pd.to_datetime(df['order_date'])
```

```
summary = df.describe()
print("\nStatistical summary:\n", summary)
median_unit_price = df['unit_price'].median()
print("\nMedian unit_price:", median_unit_price)

mean_unit_price = df['unit_price'].mean()
print("Mean unit_price:", mean_unit_price)
if mean_unit_price > median_unit_price:
    print("Observation: unit_price is right-skewed (some very high values pull the mean up).")
elif mean_unit_price < median_unit_price:
    print("Observation: unit_price is left-skewed (some very low values pull the mean down).")
else:
    print("Observation: unit_price appears symmetric.")
```
